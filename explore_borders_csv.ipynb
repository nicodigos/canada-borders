```python
df = pd.read_csv('borders.csv')
land = gpd.read_file("data/gadm41_CAN_shp/gadm41_CAN_0.shp")
land_1 = gpd.read_file("data/gadm41_CAN_shp/gadm41_CAN_1.shp")
land_2 = gpd.read_file("data/gadm41_CAN_shp/gadm41_CAN_2.shp")
land_3 = gpd.read_file("data/gadm41_CAN_shp/gadm41_CAN_3.shp")
```

```python
def plot_can_map(land): 
    land.plot(edgecolor="black", facecolor="lightblue")
    plt.title("Map of Canada")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.show()
```

```python
plot_can_map(land)
```

```python
plot_can_map(land_1)
```

```python
plot_can_map(land_2)
```

```python
plot_can_map(land_3)
```

# Check if a point is sea or land based on the level 0 shp map

```python
def is_land(latitude, longitude):
    point = Point(longitude, latitude)
    # Check if the point is within any land polygon
    return any(land.geometry.contains(point))
```

```python
lat, lon = 40.7128, -74.0060  
if is_land(lat, lon):
    print("Es tierra")
else:
    print("Es mar")
```
